```python
import csv
import pandas as pd
import scipy.stats as stats
import seaborn as sns
%matplotlib inline
```

```python
with open('census_neighborhood_demographics_2010.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    census=list(reader)
```

```python
df = pd.DataFrame(census, columns=census[0])
df.drop(0, axis=0, inplace=True)
```

### Creating demographics DataFrame and EDA

```python
demographics_df = df[['NBRHD_NAME', 'POPULATION_2010','HISPANIC_2010','WHITE_2010','BLACK_2010','NATIVEAM_2010','ASIAN_2010',
                      'HAWPACIS_2010','OTHER_2010','TWO_OR_MORE_2010','PCT_HISPANIC','PCT_WHITE','PCT_BLACK','PCT_AMERIND',
                      'PCT_ASIAN','PCT_HAW_PACIS','PCT_OTHER_RACE','PCT_TWO_OR_MORE_RACES', 'MALE', 'FEMALE']]
demographics_df.head()
```

```python
float_list = ['POPULATION_2010','HISPANIC_2010','WHITE_2010','BLACK_2010','NATIVEAM_2010','ASIAN_2010',
                      'HAWPACIS_2010','OTHER_2010','TWO_OR_MORE_2010','PCT_HISPANIC','PCT_WHITE','PCT_BLACK','PCT_AMERIND',
                      'PCT_ASIAN','PCT_HAW_PACIS','PCT_OTHER_RACE','PCT_TWO_OR_MORE_RACES', 'MALE', 'FEMALE']
for i in float_list:
    demographics_df[i] = demographics_df[i].astype(float)
```

```python
demographics_stats = demographics_df.describe().T

skew_values = list()
for num_col in demographics_stats.index:
    num_col_skew = stats.skew(demographics_df[num_col])
    skew_values.append(num_col_skew)
    
demographics_stats['skew'] = skew_values

demographics_stats
```

### Feature engineering for different types of demographics

#### adding feature for Percentage of Black population compared to mean

```python
black_upper = demographics_df['PCT_BLACK'].mean() + demographics_df['PCT_BLACK'].std()
black_lower = 0.5*demographics_df['PCT_BLACK'].mean()
black_high = demographics_df['PCT_BLACK'] > upper 
black_low = demographics_df['PCT_BLACK'] < lower
len(demographics_df[black_high]), len(demographics_df[black_low])

black_low_high = list()
for i in demographics_df['PCT_BLACK']:

    if i > black_upper:
        black_low_high.append('high')
    elif i < black_lower:
        black_low_high.append('low')
    else:
        black_low_high.append('average')
        
demographics_df['Black_PCT_cat'] = black_low_high
demographics_df.head()
```

```python
demographics_df.Black_PCT_cat.value_counts()
```

#### adding feature for Percentage of Hispanic compared to mean()

```python
hispanic_upper = demographics_df['PCT_HISPANIC'].mean() + demographics_df['PCT_HISPANIC'].std()
hispanic_lower = 0.5*demographics_df['PCT_HISPANIC'].mean()
hispanic_high = demographics_df['PCT_HISPANIC'] > hispanic_upper 
hispanic_low = demographics_df['PCT_HISPANIC'] < hispanic_lower
len(demographics_df[hispanic_high]), len(demographics_df[hispanic_low])

hispanic_low_high = list()
for i in demographics_df['PCT_HISPANIC']:

    if i > hispanic_upper:
        hispanic_low_high.append('high')
    elif i < hispanic_lower:
        hispanic_low_high.append('low')
    else:
        hispanic_low_high.append('average')
        
demographics_df['Hispanic_PCT_cat'] = hispanic_low_high
demographics_df.head()
```

```python
demographics_df.Hispanic_PCT_cat.value_counts()
```
