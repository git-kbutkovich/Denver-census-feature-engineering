### Creating DataFrame for Housing

```python
housing_df = df[['NBRHD_NAME', 'HOUSING_UNITS','OCCUPIED_HU','VACANT_HU', 'OWNED_W_MORTG_LOAN', 'OWNED_FREE_CLEAR', 
    'RENTED', 'HU_OWNED', 'HU_RENTED', 'VACANT_FOR_RENT','VACANT_RENTED_NOT_OCC','VACANT_FOR_SALE',
    'VACANT_SOLD_NOT_OCC','VACANT_SEASONAL','VACANT_OTHER']]
housing_df.head()
```

```python
for i in housing_df.columns:
    if i != 'NBRHD_NAME':
        housing_df[i] = housing_df[i].astype(float)
        
housing_df.dtypes
```

```python
housing_stats = housing_df.describe().T
housing_stats
```

```python

temp_dict = {}
def get_high_low(dataframe):
    
    for i in dataframe.columns:
   

        temp_list = list()
        if dataframe[i].dtype == float:
            temp_upper = dataframe[i].mean() + dataframe[i].std()
            temp_lower = 0.5*dataframe[i].mean()
            
            
        
            for j in dataframe[i]:
                if j > temp_upper:
                    temp_list.append('high')
                elif j < temp_lower:
                    temp_list.append('lower')
                else:
                    temp_list.append('average')
            temp_dict[i] = temp_list
    return temp_dict

       
```

```python
housing_high_low = get_high_low(dataframe = housing_df)
housing_high_low_df = pd.DataFrame(data=housing_high_low, index=list(range(1,79)))
housing_high_low_df
```
