### Looking at stats of different household makeups

```python
for i in household_df.columns:
    if i != 'NBRHD_NAME':
        household_df[i] = household_df[i].astype(float)
        
household_df.dtypes
```

```python
household_stats = household_df.describe().T

skew_values = list()
for num_col in household_stats.index:
    num_col_skew = stats.skew(household_df[num_col])
    skew_values.append(num_col_skew)
    
household_stats['skew'] = skew_values

household_stats
```

```python

temp_dict = {}
def get_high_low(dataframe):
    
    for i in dataframe.columns:
   

        temp_list = list()
        if dataframe[i].dtype == float:
            temp_upper = dataframe[i].mean() + dataframe[i].std()
            temp_lower = 0.5*dataframe[i].mean()
            
            
        
            for j in dataframe[i]:
                if j > temp_upper:
                    temp_list.append('high')
                elif j < temp_lower:
                    temp_list.append('lower')
                else:
                    temp_list.append('average')
            temp_dict[i] = temp_list
    return temp_dict

       
```

```python
high_low = get_high_low(dataframe = household_df)
high_low_df = pd.DataFrame(high_low, index=list(range(1,79)))
high_low_df

columns_cat = ['FAMILIES_cat','FAMILY_HHLD_cat','FEMALE_HHLDR_NO_CHILDR_cat','FEMALE_HHLDR_NO_HUSB_cat','FEMALE_HHLDR_W_CHILDR_cat',
               'HH_NO_NONRELATIVES_cat','HH_W_NONRELATIVES_cat','HUSB_WIFE_FAM_cat','HUSB_WIFE_NO_CHILD_cat',
               'HUSB_WIFE_W_CHLDR_cat','MALE_HHLDR_NO_CHILDR_cat','MALE_HHLDR_NO_WIFE_cat','MALE_HHLDR_W_CHLDR_cat',
               'NON_FAMILY_FEMALE_HHLDR_cat','NON_FAMILY_HH_cat','NON_FAMILY_MALE_HHLDR_cat','NUM_HOUSEHOLDS_cat',
               'ONE_PERSON_HH_cat','ONE_PERSON_HH_FEMALE_cat','ONE_PERSON_HH_MALE_cat','OTHER_FAMILY_cat','TWO_OR_MORE_PERSON_HH_cat']
```
